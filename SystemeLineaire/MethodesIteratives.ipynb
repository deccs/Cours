# Méthodes itératives

## Principe

- Matrice creuse $\mathbf{A}$ de grande taille
- Construction suite ${X^{(k)}}$ converge vers $X$ solution de $A.X=B$
- Suite d'approximations ${X^{(k)}}$ 

tq.  $\lim_{k\rightarrow\infty}{X^{(k)}}=X$

- Suite des résidues : $R^{(k)}=A.X^{(k)}-B$

tq. $\lim_{k\rightarrow\infty}{R^{(k)}}=0$
 
- **Matlab**: pgc, gmres,cgs,...
- **Comsol**: GMRES, CG, Multigrid
- **Python**: bicg, bicgstat,cg, cgs, gmres , minres

## Problème modèle: laplacien DF en 2D
$$ -\Delta u = 5\pi^2 \sin{2\pi x} \sin{\pi y}  \mbox{ dans } \Omega=[0,1]\times[0,1]$$
avec les conditions aux limites $u_\Gamma = 0$

```python
import numpy as np
from numpy.linalg import norm

def MatLap(n):
    """ matrice laplacien 2D DF """
    A=np.zeros((n*n,n*n))
    k=0
    for j in range(n):
        for i in range(n):
            A[k,k]=4.
            if i>0    : A[k,k-1]=-1.
            if i<(n-1): A[k,k+1]=-1.
            if j>0    : A[k,k-n]=-1.
            if j<(n-1): A[k,k+n]=-1.
            k=k+1
    h = 1./(n+1)
    A[:,:] *= 1./h**2
    return A
def Smb2D(n):
    """ calcul 2nd membre """
    B=np.zeros(n*n)
    h=1./(n+1)
    k=0
    for j in range(n):
        for i in range(n):
            B[k]=5*np.pi**2*np.sin(2*np.pi*h*(i+1))*np.sin(np.pi*h*(j+1))
            k=k+1
    return B
```

```python
import scipy.sparse as sps
def MatCSRLap(n):
    """ matrice laplacien 2D au format CSR """
    h = 1./(n+1)
    N=5*n**2 # nbre maxi de coef. non nul
    data=np.zeros(N)
    col =np.zeros(N,dtype=int)
    row =np.zeros(N,dtype=int)
    k=0  # numerotation des inconnues
    nz=0 # nbre de coeff. non nuls
    for j in range(n):
        for i in range(n):
            data[nz]=4./h**2
            row[nz]=k
            col[nz]=k
            nz=nz+1
            if i>0: 
                data[nz]=-1./h**2
                row[nz]=k
                col[nz]=k-1
                nz=nz+1
            if i<(n-1): 
                data[nz]=-1./h**2
                row[nz]=k
                col[nz]=k+1
                nz=nz+1
            if j>0:
                data[nz]=-1./h**2
                row[nz]=k
                col[nz]=k-n
                nz=nz+1                
            if j<(n-1): 
                data[nz]=-1./h**2
                row[nz]=k
                col[nz]=k+n
                nz=nz+1
            k=k+1
    print "construction de A dim=",n**4," non-zero=",nz," max=",N
    A=sps.csr_matrix((data[:nz],(row[:nz],col[:nz])),shape=(n**2,n**2))
    return A
```

## Méthode de Jacobi
### Principe de l'itération de Jacobi  avec $D=A_{ii}$
$$D.X^{(k+1)}=(A-D).X^{(k)}+B$$
$$X_{i}^{(k+1)}=X_{i}^{(k)}+\frac{1}{A_{ii}}(B_{i}-\sum_{j=1}^{n}A_{ij}.X_{j}^{(k)})$$
$$X_{i}^{(k+1)}=X_{i}^{(k)}+\frac{1}{A_{ii}}R_{i}^{(k)}$$
 
### condition suffisante de convergence
$A$ matrice à diagonale dominante:
$$ |A_{ii}| > \sum_{j=1, j\neq i}^n |A_{ij}| $$

### Algorithme Jacobi

    Algorithme Jacobi(A,B,N,Eps)
        # A(N,N) matrice, B(N) 2nd membre
        # N dimension, Eps précision sur le résidu
        X = 0     # initialisation solution
        répéter   # itération
            # calcul résidu Rk puis Xk
            pour i de 1 à n
                  R[i] = B[i]
                  pour j de 1 à n
                       R[i] = R[i]-A[i,j]*X[j]
                  finpour
            finpour
            pour i de 1 à n
                X[i] = X[i]+R[i]/A[i,i]
            finpour
            Residu = 0 # calcul norme résidu
            pour i de 1 à n
                Residu = Residu+R[i]*R[i]
            finpour
         jusqu'à sqrt(Residu) < Eps
         retour X

## Méthode de Gauss-Seidel 
###description de la méthode

- utilisation de $X_{i}^{(k+1)}$ dès que calculé 
- itération de Gauss-Seidel

$$X_{i}^{(k+1)}=X_{i}^{k} + \frac{1}{A_{ii}}(B_{i}-\sum_{j=1}^{i-1}A_{ij}.X_{j}^{(k+1)}-\sum_{j=i}^{n}A_{ij}.X_{j}^{(k)})$$
 

### Algorithme Gauss Seidel

    Algorithme GaussSeidel(A,B,N,Eps)
        # A(N,N) matrice, B(N) 2nd membre
        # N dimension, Eps précision sur le résidu
        X = 0     # initialisation solution
        répéter   # itération
            # calcul résidu Rk puis Xk
            pour i de 1 à n
                  R[i] = B[i]
                  pour j de 1 à n
                       R[i] = R[i]-A[i,j]*X[j]
                  finpour
                  X[i] = X[i]+R[i]/A[i,i]
            finpour
            Residu = 0 # calcul norme résidu
            pour i de 1 à n
                  R[i] = B[i]
                  pour j de 1 à n
                       R[i] = R[i]-A[i,j]*X[j]
                  finpour
                  Residu = Residu+R[i]*R[i]
            finpour
         jusqu'à sqrt(Residu) < Eps
         retour X

### Variante SOR (successive overrelaxation méthod)
paramêtre $\omega$ d'accélération de la convergence

$$X^{(k+1)}=\omega X_{GS}^{(k+1)}+(1-\omega)X^{k}$$
 

### Variante JSR Jacobi Scheduled Relaxation (Yang & Mitall JCP 2014)
- cycles de longueur M: 

pour i de 1 à M
  $$X^{(k+1)}= X^{k} + (1-\omega_i) \frac{ B - A X^{k}}{D} $$

- $\omega_i > 1 $ = cycle de sur-relaxation pour amortir les bases fréquances
- $\omega_i < 1 $ = cycle de sous-relaxation pour amortir les hautes fréquences

```python
def schedule(q,omega):
    """ repartition de omega sur le cyle JSR """
    n=len(q)
    qmax=sum(q)
    Somega=np.ones(qmax)*omega[-1]
    Somega[0]=omega[0]
    for i in range(1,n-1):
        pas = int(qmax/q[i])
        Somega[pas-1::pas]=omega[i]
    return Somega

def cycle(level,n):
    """ détermination du cycle JSR fonction du niveau level
        et de n avec k_min=O(1/n) val propre de A
    """
    if level == 1:
        return [1],[1.0]
    elif level == 2:
        if n>=128:
            return [1,130],[425.8,0.9742]
        elif n>=64:
            return [1,63],[190.2,0.9532]
        else:
            return [1,15],[32.6,0.8630]
    elif level == 3:
        if n>=256:
            return [1,27,625],[6319,63.99,0.8989]
        elif n>=64:
            return [1,11,106],[684.3,20.73,0.8149]
        else:
            return [1,5,21],[64.66,6.215,0.7042]        
    elif level == 4:
        if n>=256:
            return [1,9,86,664],[12329.,492.05,15.444,0.78831]
        elif n>=64:
            return [1,5,26,114],[1029.4,95.007,6.3913,0.70513]
        else:
            return [1,2,8,20],[80.154,17.217,2.6201,0.62230]
    else:
        print "niveau incorrect ",level
        return None,None

def Jacobi(A,B,level,eps):
    """ resolution AX=B avec Jacobi JSR  et A sparse"""
    N = A.shape[0]    
    #X = np.random.rand(N)
    X = np.zeros(N)
    D = A.diagonal()
    itmax=10*N  
    Err= np.zeros(itmax)
    # parametres des relaxations JACOBI
    q,omega=cycle(level,int(np.sqrt(N)))
    Somega = schedule(q,omega)
    qmax = sum(q) # nbre d'it / cycle
    print "cycle ",len(q)," M=",q,qmax
    print "OMEGA=",omega
    # lissage de la solution initiale
    nit = 0
    R = B.copy() 
    err0 = norm(R)
    err  = 1.0
    # debut boucle
    while err>eps:
        # sous iteration de relaxation
        for it in range(qmax):
                X[:] += Somega[it]*(R/D)
                R[:]  = B - A.dot(X)
                err = norm(R)/err0
                Err[nit]=err
                nit = nit +1
                if err<eps: break
        if nit>=(itmax-qmax): break
    return X,Err[:nit]
```

```python
n=128
print "Probleme laplacien N=",n**2,"(",n,")"
B=Smb2D(n)
A=MatCSRLap(n)
```

```python
print "dimension 1D n=",n
eps=10e-9
level=4
X1,Err1=Jacobi(A,B,1,eps)
rho1=len(Err1)*np.log(0.1)/np.log(Err1[-1])
print "nit=",len(Err1)," err=",Err1[-1]," cvge=",int(rho1)
X2,Err2=Jacobi(A,B,level,eps)
rho2=len(Err2)*np.log(0.1)/np.log(Err2[-1])
print "nit=",len(Err2)," err=",Err2[-1]," cvge=",int(rho2)
print "rho(JSR)=",rho1/rho2
#X2,Err2=GaussSeidel(A,B,1.e-012)
plt.semilogy(np.arange(len(Err1)),Err1,label='Jacobi')
plt.semilogy(np.arange(len(Err2)),Err2,label='Jacobi JSR')
plt.legend(loc=0)
plt.show()
```

# Méthode de gradient

## description de la méthode

- A est une matrice symétrique définie positive si 
$$X^{t}.A.X > 0 \mbox{ i.e. } \lambda_{i}>0 $$
 
- soit la forme quadratique définie positive 
$$ \phi(X)=\frac{1}{2}X^{t}.A.X-X^{t}.B $$
$$ \phi(X)=\sum_{i=1}^{n}\sum_{j=1}^{n}X_{j}.A_{ij}.X_{i}-\sum_{i=1}^{n}X_{i}.B_{i}$$
 
- Le minimum de $\phi(X)$ est solution de $A.X=B$
 $$ X = Arg Min(\phi(X)) \Leftrightarrow A.X=B$$
 $$\frac{\partial\phi(X)}{\partial X_{i}}=\sum_{j=1}^{n}A_{ij}.X_{j}-B_{i}$$
 
 ### Minimisation de $\phi(X)$
  
   - construction suite $X^{(k)}$ tq $\phi(X^{(k+1)})<\phi(X^{k})$
   
   - choix d'une direction de descente  $R^k$
  $$R^{(k)}=B-A.X^{(k)}=-\overrightarrow{grad}(\phi(X^{(k)}))$$
  
   - minimisation dans cette direction i.e. minimisation de $\phi(X^{k}+\lambda R^{k})$ par rapport à $\lambda$, i.e.
  $$\frac{d}{d\lambda}\left(\frac{1}{2}(X^{k}+\lambda R^{k})^t A (X^{k}+\lambda R^{k}) - (X^{k}+\lambda R^{k})^t B\right)=0$$
  pour 
  $$\lambda=\frac{R^{(k)}.R^{(k)}}{R^{(k)}.A.R^{(k)}}$$


```python
from numpy.linalg import solve
N=4; N2=N*N
B=Smb2D(N)
A=MatLap(N)
Xe = solve(A,B)
Phi = lambda X: 0.5*np.dot(X,np.dot(A,X))-np.dot(B,X)
# point initial
X0 = Xe + 0.5*(2*np.random.rand(N2)-1.)
# direction gradient
R0 = B-np.dot(A,X0)
V0 = R0/norm(R0)
```

```python
def pltPhi2D(X0,V1,V2):
    """tracer de phi dans la plan V1,V2 passant par X0"""
    N=len(X0)
    n=21
    s=np.linspace(-1,1,n)
    phi=np.zeros((n,n))
    for i,s1 in enumerate(s):
        for j,s2 in enumerate(s):
            phi[i,j]=Phi(X0+s1*V1+s2*V2)
    x0=n/2; y0=n/2;
    phi0=Phi(X0)
    plt.contour(phi,level=[phi0])
    c=plt.contour(phi)
    plt.clabel(c)
    plt.plot([x0],[y0],'c*',ms=10)
    return
def pltPhi1D(X0,V0,beta):
    """ tracer phi suivant V0 """
    N=len(X0)
    n=21
    s=np.linspace(-2*beta,2*beta,n)
    phi=np.zeros(n)
    for i,s1 in enumerate(s):
        phi[i] = Phi(X0+s1*V0)
    plt.plot(s,phi)
    phi0=Phi(X0)
    plt.plot(0.,phi0,'c*',ms=10)
    phi1=Phi(X0+beta*V0)
    plt.plot(beta,phi1,'ro',ms=10)
    return
```

```python
# directions de projection
V1 = np.zeros(N2); V1[0]=1.0;  V1 = V1/norm(V1)
V2 = np.zeros(N2); V2[1]=1.0;  V2 = V2/norm(V2)
V3 = np.zeros(N2); V3[3]=1.0;  V3 = V3/norm(V3)
```

```python
print "Phi(X0)=%g  Phi(Xe)=%g"%(Phi(X0),Phi(Xe))
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.axis('equal')
pltPhi2D(X0,V1,V2)
plt.subplot(1,2,2)
plt.axis('equal')
pltPhi2D(X0,V3,V1)
plt.draw()
```

```python
# minimisation suivant le gradient
beta = np.dot(R0,R0)/np.dot(R0,np.dot(A,R0))
X1 = X0 + beta*R0
print "Phi0=%g beta=%g Phi1=%g"%(Phi(X0),beta,Phi(X1))
plt.figure()
pltPhi1D(X0,R0,beta)
plt.draw()
```

### Itération de gradient
 - tant que $|R^k|>\epsilon$
  $$X^{(k+1)}=X^{(k)}+\beta_{k}.R^{(k)}\mbox{ avec }\beta_{k}=\frac{R^{(k)}.R^{(k)}}{R^{(k)}.A.R^{(k)}}$$
  $$ R^{(k+1)} = B - A X^{(k+1)} $$ 
  

```python
eps=1.e-6
def MethodeGradient(A,B,X0):
    itmax=3*len(X0)
    Err=np.zeros(itmax)
    X=X0.copy()
    R=B-np.dot(A,X)
    R2=np.dot(R,R)
    it = 0
    while np.sqrt(R2)>eps:
        beta=R2/np.dot(R,np.dot(A,R))
        X = X + beta*R
        R = B - np.dot(A,X)
        R2 = np.dot(R,R)
        Err[it]=np.sqrt(R2)
        it = it+1
        if it>=itmax: break
    return X,Err[:it]
```

## convergence de la méthode

**théorème:** Si A est une matrice à symétrique définie positive, alors la méthode de gradient converge 

**demonstration**
$$\phi(X^{(k+1)})=\min_{\lambda\in\mathbb{R}}\phi(X^{(k)}+\lambda.R^{(k)})$$
car $\frac{\partial\phi(\lambda_{k})}{\partial\lambda}=0$
et donc $\phi(X^{(k+1)})<\phi(X^{(k)})$

d'ou la convergence de l'algorithme de gradient

```python
X,Err=MethodeGradient(A,B,X0)
nit=len(Err)
print "err(X0) =",norm(X0-Xe)
print "Convergence en %d iterations"%(nit)
print "err(Xk) = ",norm(X-Xe)
x=np.arange(nit)
plt.semilogy(x,Err)
plt.title("Methode de gradient")
plt.show()
```

## Méthode de gradient conjugués

methode de gradient: $X^{(k+1)} = X^{(k)} + \beta_k Q^{(k+1)}$

avec une direction de descente $Q^{(k+1)}$  orthogonale à la direction précédente $Q^{(k)}$:

$$ \left(Q^{(k)}\right)^{t} A Q^{(k+1)} =0 \mbox{ avec } Q^{(k+1)}=R^{(k)}+\beta_{k} Q^{(k)} $$
 

d'où:

$$\beta_{k}=-\frac{\left(P^{(k)}\right)^{t}.\left(AR^{(k)}\right)}{\left(P^{(k)}\right)^{t}.\left(AP^{(k)}\right)}$$
 

L'itération de gradient conjugué s'écrit:

$$X^{(k+1)}=X^{(k)}+\lambda_{k}.P^{(k)}\mbox{ avec }\lambda_{k}=\frac{\left(P^{(k)}\right)^{t}.\left(R^{(k)}\right)}{\left(P^{(k)}\right)^{t}.\left(A.P^{(k)}\right)}$$
 

La convergence de la méthode est quadratique, i.e. en $\theta(\sqrt{N})$

**remarque** méthode directe car convergenve en au plus N itérations !!!

- **Python** scipy sparse linalg cg 

```python
from scipy.sparse.linalg import cg
itmax=4*len(B)
Err=np.zeros(itmax)
nit=0
def CalcErr(Xk):
    global A,B,Err,nit
    Err[nit]=norm(B-np.dot(A,Xk))
    nit=nit+1
    return
# resolution par GC
X,code=cg(A,B,x0=X0,tol=eps,maxiter=itmax,callback=CalcErr)
# bilan
print "err(X0) =",norm(X0-Xe)
print "Convergence en %d iterations"%(nit)
print "err(Xk) = ",norm(X-Xe)
x=np.arange(nit)
plt.semilogy(x,Err[:nit])
plt.title("Methode de gradient conjugue")
plt.show()
```

## Influence du conditionnement

La convergence de la méthode de gradient dépend du conditionnement de la matrice A. 

Plus le conditionnement augmente, et plus la vitesse de convergence diminue.

Pour améliorer la convergence, on préconditionne le système en multipliant par une matrice de préconditionnement $P^{-1}$:

$$A.X=B\,\leadsto\,\,\, P^{-1}A\, X=P^{-1}B$$
 

telle que $cond(M^{-1}A)\approx 1$, i.e. que $P^{-1}$ ait le même spectre de valeurs propres que $A^{-1}$ ($P^{-1}\approx A^{-1}$). 

**remarque**: on ne calcule jamais explicitement $P^{-1}$, mais on résoud des systèmes linéaires $P.Y=V$ pour calculer $Y=M^{-1} V$
 

**remarque**: le préconditionnement peut être fait à l'aide d'une méthode itérative (Jacobi, Gauss Seidel) ou par des factorisations incomplètes (ILU, Incomplet Cholesky). 


```python
from scipy.sparse.linalg import spilu,LinearOperator
n=len(B)
P=spilu(A.tocsc(),fill_factor=10.)
print "Preconditionnement P nnz=",P.nnz," A nnz=", A.nnz
M_x = lambda x: P.solve(x)
M = LinearOperator((n,n),M_x)
# resolution par GC preconditione
nit=0
U,code=cg(A,B,M=M,tol=eps,maxiter=itmax,callback=CalcErr)
x1p=np.arange(nit)
Err1p=Err[:nit].copy()
# bilan
print "err(X0) = ",Err[0]
print "Convergence en %d iterations"%(nit)
print "err(Xk) = ",Err[nit-1]
plt.semilogy(x1p,Err1p,label='cg prec',lw=2)
plt.semilogy(x1,Err1,label='cg')
plt.legend()
plt.title("Methode de gradient conjugue preconditionne")
plt.show()
```

## Méthode de Krylov

**espace de Krylov:** on choisit un vecteur $v_{1}=b/|b|$, et on construit le sous-espace de Krylov $H_{m}=\{v_{1},A.v_{1},A^{2}v_{1},...A^{m-1}v_{1}\}$. 

- Une base orthogonale de $H_{m}$ est une bonne approximation des $m$ plus grands vecteurs propres de $A$

- remarque pour m=n on a $H_{n}=\mathcal{R}^{n}$. 

**résolution itérative** de $Ax=b$ avec A non singulière de dimension n.

**principe**: soit $x_{0}$ un vecteur initial, on calcule une nouvelle estimation $x_{1}$ telle que le résidu $r_{1}=b-A.x_{1}$ soit minimale dans l'espace de Krylov $H_m$. 

On détermine une base orthogonale $\left(q^{k}\right)_{k=1,m}$ de $H_m$ par la méthode d'Arnoldi, qui est plus stable que l'orthogonalisation de Gram-Schmidt

- **Algorithme d'Arnoldi**
   1. Start with an arbitrary vector q1 with norm 1.
   2. Repeat for k = 2, 3, …
   
     - $q_k \leftarrow Aq_{k-1} $
     
     - for j from 1 to k − 1
         - $h_{j,k-1} \leftarrow q_j^* q_k $
         - $q_k \leftarrow q_k - h_{j,k-1} q_j$
     - $h_{k,k-1} \leftarrow \|q_k\|$
     - $q_k \leftarrow \frac{q_k}{h_{k,k-1}}$

Soit $Q^m$ la matrice (n,m) formée par les $q^k$, on a alors
$$x_1 = Q^m y_1 \mbox{ avec } y_1\in R^m$$




 
$y_1$ est alors obtenue en minimisant un problème de moindre carré d'ordre $m$

$$\min | H y_1 -\beta e_1| \mbox{ avec } \beta =|b-A.x_1|$$

où $H$ est la matrice de Hessenberg tq $A Q^n = Q^{n+1} H$
 

**Algorithme GMRES**

1. calcul $r_{0}=b-Ax_{0}$
 
2. génère une base $\{q_{k}\}$ de $H_{m}$  par un procédé d'Arnoldi

3. calcul correction $y_{m}=argmin_{y}||\beta e_{1}-H y||$
 
4. $x_{m}=x_{0}+Q y_{m}$
 
5. faire $x_{0}=x_{m}$ et allez en 1

```python
from scipy.sparse.linalg import gmres
eps=1.e-10
itmax=4*len(B)
Err=np.zeros(itmax)
nit=0
def CalcErr(Rk):
    global Err,nit
    Err[nit]=norm(Rk)
    nit=nit+1
    return
# resolution par GMRES 
nit=0
U,code=gmres(A,B,tol=eps,restart=10,maxiter=itmax,callback=CalcErr)
x2=np.arange(nit)
Err2=Err[:nit].copy()
nit=0
U,code=gmres(A,B,tol=eps,restart=50,maxiter=itmax,callback=CalcErr)
x3=np.arange(nit)
Err3=Err[:nit].copy()
# bilan
print "err(X0) = ",Err[0]
print "Convergence en %d iterations"%(nit)
print "err(Xk) = ",Err[nit-1]
x=np.arange(nit)
plt.semilogy(x2,Err2,label='gmres n=10')
plt.semilogy(x3,Err3,label='gmres n=50')
plt.semilogy(x1,Err1,label='cg')
plt.legend()
plt.title("Methode gmres")
```

## Préconditionement
Pour améliorer la convergence, on préconditionne le système en effectuant par exemple une factorisation incompléte de $A$:

$$A=L.U+R$$
où $L$ et $U$ ont la même structure creuse que $A$.

Le système préconditionné s'écrit:

$$A.x=B\,\leadsto\,\,\, L^{-1}AU^{-1}\, y=L^{-1}b\,\,\, avec\,\,\, x=U^{-1}y$$

# Méthode Multigrille
résolution système linéaire $N\times N$ en $\theta(N logN)$ itérations
(mais pour $N$ très grand)

## Principe

On considère le problème de Poisson monodimensionnel avec des conditions aux limites de Dirichlet homogènes:

$$-\frac{d^{2}u}{dx^{2}}=f\,\,\,\,\,\mbox{ sur }\,\,\Omega=]0,1[\,\,\,\mbox{ avec }\, u(0)=u(1)=0$$
 
Pour exposer le plus simplement possible le principe de la méthode multigrille , on commence par présenter la méthode 2-Grilles.

Cette méthode combine deux méthodes peu performantes pour finalement obtenir une méthode très efficace.


### Méthode 2-grilles
 On considère une grille fine de pas h et une grille grossière de pas H = 2h pour résoudre le problème 
 
Un cycle de la méthode 2-Grilles se compose de deux phases :

- une méthode de **lissage** correspondant à deux ou trois itérations d'une méthode de relaxation (GaussSeidel, Jacobi), permettant de réduire les hautes fréquences de l'erreur lorsque l'on décompose celle-ci dans la base des vecteurs propres de la matrice; 

soit $u_{h}$ l'approximation de $U=U_{h}$ solution du système linéaire discret $A U_{h}=F_{h}$
 

- une méthode de **correction** sur grille grossière qui traite efficacement les basses fréquences de l'erreur . La correction $V_{h}=U_{h}-u_{h}$, qu'il faut ajouter à $u_{h}$ (approximation) pour obtenir $U_{h}$ (solution exacte), doit vérifier :

$$A\, V_{h}=r_{h}\,\,\,\mbox{ avec }\, r_{h}=F_{h}-A\, u_{h}$$
 
$V_{h}$ est donc la solution d'un problème du même type que celui qui définit $U_{h}$, où le second membre $F_{h}$ a été remplacé par le résidu $r_{h}$ aux points de la grille fine. 

Le calcul de $V_{h}$ est a priori aussi coûteux que celui de $U_{h}$

- Dans la mesure où l'erreur est lissée, on peut cependant chercher à obtenir une approximation de $V_{h}$ sur la grille grossière de pas $H=2h$, ce qui nécessitera beaucoup moins de calculs puisque la grille de pas $H$ possède environ deux fois moins de points dans le cas du problème monodimensionnel (quatre fois moins, dans le cas du problème bidimensionnel, huit fois moins dans le cas du problème tridimensionnel) que celle de pas $h$

- on pourra effectivement obtenir une bonne approximation de $V_{h}$ ur la grille grossière si $V_{h}$ varie lentement, ce qui est effectivement le cas grâce aux itérations de lissage. 

La seconde phase de chaque cycle de la méthode 2-Grilles consiste alors à :

1. définir la restriction $r_{H}$ de $r_{h}$ sur la grille grossière à l'aide d'un opérateur de restriction; cet opérateur peut être, tout simplement, l'injection définie en tous points M de la grille grossière, mais on peut également utiliser une restriction par moyenne pondérée ;

2. résoudre le système linéaire $AV_{H}=r_{H}$, donnant la correction $V_{H}$ sur la grille grossière ;

3. prolonger $V_{H}$ obtenue sur la grille grossière en une fonction obtenue sur la grille fine, par exemple par interpolation bilinéaire.

4. On recommence ensuite un nouveau cycle de la méthode 2-Grilles

La question principale qui se pose est de savoir si, effectivement, le prolongement de $V_{H}$ est une bonne approximation de $V_{h}$ et, dans ce cas, quel est le gain que l'on va attendre de l'utilisation de la méthode 2-Grilles ; il est à noter que cette méthode 2-Grilles peut s'écrire sous la forme d'une itération de point fixe linéaire du type 

$$U_{k+1}=B\, U_{k}+C\,\,\mbox{ pour }\,\, k=0,1,2,..$$
 

Pour un lissage correspondant à la méthode de Gauss-Seidel, on montre, par analyse de Fourier, que le facteur de réduction de l'erreur est égal à $\sqrt{5}/5\approx 0,447$ ce qui correspond à un excellent taux de réduction.

## Méthode multigrille

La méthode multigrille correspond à l'application récursive d'une méthode 2-Grilles. En effet lors de la résolution du système d'équations sur la grille grossière $AV_{H}=r_{H}$ , on peut de nouveau utiliser une méthode 2-Grilles, en définissant un problème sur une grille encore plus grossière de pas 2H , et ainsi de suite pour finalement résoudre un système d'équations sur la grille la plus grossière ne contenant que quelques points, voire même un seul point puisque les pas de discrétisation sont choisis comme des puissances inverses de 2. 

Cela nécessite de définir des stratégies de passages intergrilles. Signalons également que cette méthode multigrille est bien adaptée à une utilisation sur des domaines rectangulaires ; il existe cependant des variantes de la méthode multigrille bien adaptées à la résolution numérique d'équations aux dérivées partielles sur des domaines quelconques .

## Exemple
Soit la méthode de Jacobi pour résoudre

\begin{align}
u''(x) & = f(x) & 0<x<1
\end{align}

avec $u(0) = 1, u(1) = 3$ et

\begin{align}
f(x) & = -20 + \frac{1}{2} \phi''(x) \cos(\phi(x)) - \frac{1}{2} (\phi'(x))^2\sin(\phi(x)) \\ \\
\phi(x) & = 20\pi x^3.
\end{align}

```python
m=2**8-1                           # Number of grid points
h=1./(m+1)                         # Mesh width
x=np.linspace(0,1,m+2); x=x[1:-1]  # grid
alpha=1.; beta=3.                  # boundary values
phi = lambda x: 20.* np.pi * x**3
f = lambda x: -20 + 0.5*120*np.pi*x * np.cos(phi(x)) - \
                0.5*(60*np.pi*x**2)**2 * np.sin(phi(x)) # RHS
plt.plot(x,f(x)); plt.xlabel('x'); plt.ylabel('f(x)')
plt.title("Second membre f(x)")
plt.show()
```

## Méthode de Jacobi
schéma différences finies centré

\begin{align}
\frac{U_{i-1} - 2 U_i + U_{i+1}}{h^2} & = f(x_i).
\end{align}

soit pour  $U_i$
\begin{align}
U_i & = \frac{1}{2} ( U_{i+1} + U_{i-1} ) - \frac{h^2}{2} f(x_i).
\end{align}

Méthode de Jacobi = itération de points fixe:

\begin{align}
U_i^{[k+1]} & = \frac{1}{2} ( U_{i+1}^{[k]} + U_{i-1}^{[k]} ) - \frac{h^2}{2} f(x_i).
\end{align}

Calcul la solution sur 100 iterations, puis la trace

```python
U=np.linspace(alpha,beta,m)       # Just use a straight line as initial guess
UU=[U]
F=0.5*h**2*f(x)                   # Construct the RHS
F[0]-=alpha/2.; F[-1]-=beta/2.    # boundary conditions
e=np.ones(m-1)
G=0.5*(np.diag(e,-1)+np.diag(e,1)) # Jacobi matrix (average neighbors)
for i in range(100):
    U=np.dot(G,U) - F
    UU.append(U)
```

## restriction et prolongation
- transfert sur la grille grossière: coarsen
- interpolation sur la grille fine: interpolate


```python
def coarsen(f):
    return f[1::2]  # This slicing just takes the odd-numbered points

def interpolate(f,alpha,beta):
    m_coarse=len(f)
    m_fine  =2*m_coarse+1
    f_interp = np.zeros(m_fine)
    f_interp[1::2]=f                          #Set the values of the odd numbered points
    f_interp[2:-1:2]=0.5*(f[:-1]+f[1:])       #Set the values of the (interior) even numbered points
    f_interp[0]=0.5*(f_interp[1]+alpha)     #Set the values of the endpoints
    f_interp[-1]=0.5*(f_interp[-2]+beta)
    return f_interp
```

- puis interpolation de la solution sur la grille fine.

```python
m=2**8-1                           # Number of grid points
h=1./(m+1)                         # Mesh width
x=np.linspace(0,1,m+2); x=x[1:-1] 
u = 1.+12.*x-10.*x**2 + 0.5*np.sin(phi(x))  # Exact solution on computational grid
omega=2./3.
print "Maillage ",m," omega=",omega
F=0.5*h**2*f(x)                   # Construct the RHS
F[0]-=alpha/2.; F[-1]-=beta/2.    # boundary conditions
# iteration grille fine
U1=np.linspace(alpha,beta,m)
e=np.ones(m-1)
G=0.5*(np.diag(e,-1)+np.diag(e,1)) # Jacobi matrix (average neighbors)
for i in range(3):
    U1=(1.-omega)*U1+omega*(np.dot(G,U1) - F)
#compute the residual
U=U1.copy()
A=2./h**2*(G-np.eye(m))
r=F-np.dot(A,U)
print "Residu init ",max(abs(r))
m2=(m-1)/2.               # Number of points in coarse grid
h2=1./(m2+1)              # coarse mesh width
x2=coarsen(x)             # coarse grid points
r2=coarsen(r)             # residual restricted to coarse grid
# calcul iteratif
if True:
    err2=np.zeros(m2)     # initial guess for the error
    e2=np.ones(m2-1); 
    G2=0.5*(np.diag(e2,-1)+np.diag(e2,1))
    for i in range(3):    #Solve Ae=-r by Jacobi iteration
        err2=(1.-omega)*err2+omega*(np.dot(G2,err2)+0.5*h2**2*r2)
else:
    # calcul exacte
    A2=2./h2**2*(G2-np.eye(m2))
    # resolution 
    err2=solve(A2,-r2)
# interpolation
err=interpolate(err2,0,0)     # interpolate the error
print "Err ",max(abs(err))
U=U-err                   # and use it to correct our solution
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(x,u,x,U,x,U-u)
plt.legend(['Exact','Jacobi','Error'],loc='best')
plt.subplot(1,2,2)
plt.plot(x,U1-u,x,U-u)
plt.legend(['Jacobi','2grille'],loc='best')
```

## V cycle multigrille

- k niveaux de grilles

```python
def Jacobi(U,f,alpha,beta,m,nu):
    """Perform nu Jacobi iterations on a grid with m points, with initial guess U, right hand side function f and 
       Dirichlet boundary conditions with values alpha and beta.  The function returns both the approximate
       solution and the residual."""
    omega=2./3.
    h=1./(m+1)
    F=0.5*h**2*f.copy()
    F[0]-=alpha/2.; F[-1]-=beta/2.
    e=np.ones(m-1)
    G=0.5*(np.diag(e,-1)+np.diag(e,1))
    for i in range(nu):
        U=(1.-omega)*U + omega*(np.dot(G,U)-F)
    A=2./h**2*(G-np.eye(m))
    FF=f.copy(); FF[0]-=alpha/h**2; FF[-1]-=beta/h**2
    rr=FF-np.dot(A,U)
    return U,rr
```

```python
# nbre grilles k
k=8;
nu=3      # Number of Jacobi iterations to take at each step
m=2**k-1
x=np.linspace(0,1,m+2); x=x[1:-1]  # grid
phi = lambda x: 20.* np.pi * x**3
u = 1.+12.*x-10.*x**2 + 0.5*np.sin(phi(x))
f = lambda x: -20 + 0.5*120*np.pi*x * np.cos(phi(x)) - 0.5*(60*np.pi*x**2)**2 * np.sin(phi(x))
alpha=1.; beta=3.
F=f(x)
# VCycle profondeur rdep
def Vcycle(rdep):
    m=2**k-1
    if rdep>k:
        rdep=k-1  # Recursion depth; this is how many grids down we want to go
                  # rdep=k-1 gives a full V-cycle
    U=np.linspace(alpha,beta,m)  # Initial guess
    r=[None]*(rdep+1); error=[None]*(rdep+1) # This just initializes these lists to have the right length
    res=[None]*(2*rdep+1)
    U,rr=Jacobi(U,F,alpha,beta,m,nu)         # Initial iteration on fine grid
    res[0]=rr.copy()
    # V cycle descente 
    for i in range(1,rdep+1): # Going down
        m=(m-1)/2 #  = 2**(k-i) - 1
        r[i]=coarsen(rr)             # residual restricted to next coarser grid
        error[i],rr=Jacobi(np.zeros(m),-r[i],0.,0.,m,nu)
        res[i]=rr.copy()
    # V cycle remonter
    for i in range(1,rdep): # Coming up
        m=2*m+1
        err=error[rdep-i]-interpolate(error[rdep+1-i],0,0)     # Interpolate and subtract the correction
        error[-i-1],rr=Jacobi(err,-r[rdep-i],0.,0.,m,nu)
        res[rdep+i]=rr.copy()
    m=2*m+1
    U=U-interpolate(error[1],0,0)                   # final solution correction
    U,rr=Jacobi(U,F,alpha,beta,m,nu)                #Final iterations on original grid
    res[-1]=rr.copy()
    # tracer
    plt.figure(figsize=(12,4))
    plt.subplot(1,3,1)
    plt.plot(x,u,x,U,'-',x,U-u)
    plt.legend(['Exact','Jacobi','Error'],loc='best')
    plt.subplot(1,3,2)
    plt.plot(res[0])
    plt.plot(res[-1])
    plt.subplot(1,3,3)
    nres=[norm(r) for r in res]
    plt.semilogy(nres,'-o')
    plt.show()
    return
```
